### 24년 bat WAR 반복문 실행

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
from urllib.request import Request

club_dict = { "2002" : "KIA", "12001" : "KT", "10001" : "키움", "11001" : "NC", "5002" : "LG",
              "1001" : "삼성", "6002" : "두산", "9002" : "SSG", "3001" : "롯데", "7002" : "한화"}

bat_war_name_24 = []
bat_club_name_24 = []
bat_war_score_24 = []
bat_war_game_24 = []

page_source = driver.page_source
soup = BeautifulSoup(page_source, "html.parser")

bat_war_list_main_class = soup.find("tbody")
#bat_war_list_sub_class = bat_war_list_main_class.find("div","table_type01.transverse_scroll.cbox")
bat_war_list = bat_war_list_main_class.find_all("tr")

start_indices = list(range(2, len(bat_war_list), 12))

for start in start_indices:
    end = start + 10  # 각 그룹은 10개의 데이터를 포함
    if end > len(bat_war_list):
        end = len(bat_war_list)  # 데이터 리스트의 길이를 초과하지 않도록 조정

    for bat_war_list_idx in range(start, end):
        bat_war_name_24.append(bat_war_list[bat_war_list_idx].find("a").getText())  # 이름 취득

        bat_war_club_list = str(bat_war_list[bat_war_list_idx].find_all("img", src=True))  # 클럽 정보 취득
        bat_war_club_code = bat_war_club_list.split('/')[-2].split('.')[0]  # 클럽 코드 취득
        bat_club_name_24.append(club_dict.get(bat_war_club_code))  # 클럽 이름 취득

        bat_war_score_list = bat_war_list[bat_war_list_idx].find_all("td")
        bat_war_score_24.append(bat_war_score_list[3].get_text())  # WAR 데이터 취득

        bat_war_game_24.append(bat_war_score_list[6].get_text())  # WAR 데이터 취득



print(bat_war_name_24)
print(bat_club_name_24)
print(bat_war_score_24)
print(bat_war_game_24)
```
