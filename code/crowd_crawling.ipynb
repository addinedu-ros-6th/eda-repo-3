```python
from selenium.webdriver import ActionChains
import time
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen
from urllib.request import Request

action = ActionChains(driver)
page_number = True
page_icon_shift = 0
first_page = 2
ranking_22 = []
name_22 = []
team_22 = []
agv_22 = []
g_22 = []
pa_22 = []
ab_22 = []
h_22 = []
double_hit_22 = []
triple_hit_22 = []
hr_22 = []
rbi_22 = []
sb_22 = []
cs_22 = []
bb_22 = []
hbp_22 = []





while page_number == True:
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, "html.parser")  

    page_list_main_class = driver.find_element(By.CLASS_NAME,"record_result")
    page_list_sub_class = page_list_main_class.find_element(By.CLASS_NAME, "paging")
    page_list = page_list_sub_class.find_elements(By.TAG_NAME,"a")


    page_soup_main_class = soup.find("div","record_result")
    page_soup_sub_class = page_soup_main_class.find("div","paging")
    page_number_list = page_soup_sub_class.find_all("a")

    page_number_tmp = page_number_list[len(page_number_list)-2].getText()
    if page_number_tmp == '':
        page_offset = 1
    else:
        page_offset = 0

    for page_idx in range(first_page,len(page_list)-page_offset):
        last_height = driver.execute_script("return document.body.scrollHeight")
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        page_source = driver.page_source
        soup = BeautifulSoup(page_source, "html.parser")
    
        runner_rcd_main_class = soup.find("table", "tData01 tt")
        runner_rcd_sub_class = runner_rcd_main_class.find("tbody")
        runner_rcd_list = runner_rcd_sub_class.find_all("tr")

        for player_idx in range(len(runner_rcd_list)):


            player_rcd_runner_raw = runner_rcd_list[player_idx].getText()
            player_rcd_runner = player_rcd_runner_raw.split("\n")
            ranking_22.append(player_rcd_runner[1])
            name_22.append(player_rcd_runner[2])
            team_22.append(player_rcd_runner[3])
            agv_22.append(player_rcd_runner[4])
            g_22.append(player_rcd_runner[5])
            pa_22.append(player_rcd_runner[6])
            ab_22.append(player_rcd_runner[7])
            h_22.append(player_rcd_runner[8])
            double_hit_22.append(player_rcd_runner[9])
            triple_hit_22.append(player_rcd_runner[10])
            hr_22.append(player_rcd_runner[11])
            rbi_22.append(player_rcd_runner[12])
            sb_22.append(player_rcd_runner[13])
            cs_22.append(player_rcd_runner[14])
            bb_22.append(player_rcd_runner[15])
            hbp_22.append(player_rcd_runner[16])
            time.sleep(0.1)
            print(player_rcd_runner)

        page_list_main_class = driver.find_element(By.CLASS_NAME,"record_result")
        page_list_sub_class = page_list_main_class.find_element(By.CLASS_NAME, "paging")
        page_list = page_list_sub_class.find_elements(By.TAG_NAME,"a")
        first_page,len(page_list)
        page_list[page_idx].click()  # 페이지 넘기기
        

    
        time.sleep(0.5)
    first_page = 3


    print(page_number_tmp)
    if page_number_tmp == '':
        page_number = True
    else :
        page_number = False
```

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
from urllib.request import Request


page_source = driver.page_source
soup = BeautifulSoup(page_source, "html.parser")

sql = ('''insert into crowd values(%s, %s, %s, %s, %s, %s)''')

crowd_main_class_list = soup.find("tbody")
crowd_list = crowd_main_class_list.find_all("tr")


for crowd_idx in range(len(crowd_list)):
    crowd = crowd_list[crowd_idx].getText().strip().split('\n')
    crowd[5]= int(crowd[5].replace(',',''))
    value = (crowd[0],crowd[1],crowd[2],crowd[3],crowd[4],crowd[5])

    kbo_base.execute(sql,tuple(value))

    print(tuple(value))

remote.commit()

remote.close()

```
